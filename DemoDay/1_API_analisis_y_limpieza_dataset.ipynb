Primero se identifican las columnas, la cantidad de filas sin nulos y los tipos de datos de cada columna

```python
df.info()
```

Ahora se puede obtener una descripción estadistica de los datos numéricos

```python
df.isna()
```

los NaN encontrados por columna son:

```python
df.isna().sum(axis=0) # Los NaN encontrados por fila son: df.isna().sum(axis=1)
```

Se eliminan las filas donde todos los valores sean NaN

```python
df_limpio_na = df.dropna(axis=0, how='all')
df_limpio_na 
```

Para los registros que no tengan max_supply voy a actualizar el valor con 0

```python
df_limpio_na['max_supply'] = df_limpio_na['max_supply'].fillna(0)
df_limpio_na
```

```python
df_limpio_na.describe() # Descripción estadistica de los datos numéricos
```

Para tener un historico de los datos diarios, se almacena el dataframe en un archivo CSV en el cual se pueda ver la fecha en el que se descargaron los datos

```python
fecha= datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
archivo_fecha=f"divisas{fecha}.csv"
df_limpio_na.to_csv(archivo_fecha)
```
