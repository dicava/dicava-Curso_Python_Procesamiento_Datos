Vamos a encontrar la divisa que ha tenido mayor variación con respecto al ranking

```python
max_delta=max(df_2.rank_delta)
min_delta=min(df_2.rank_delta)
max_delta
min_delta

def mayor_variacion(a, b):
    if abs(a)>abs(b):
        return a
    else:
        return b
    
mayor_var=mayor_variacion(max_delta, min_delta)
mayor_var
```
